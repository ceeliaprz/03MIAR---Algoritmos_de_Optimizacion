##Torres de Hanoi con Divide y vencerás

```
def Torres_Hanoi(N, desde, hasta): #divide y vencerás va reduciendo el problema haciendo N-1
  colores = {1: "Roja", 2: "Amarilla", 3: "Verde"}
  if N ==1 :
    print("Lleva la ficha ", colores[N], "desde", desde , " hasta ", hasta )

  else:
    #Torres_Hanoi(N-1, desde, 6-desde-hasta )
    Torres_Hanoi(N-1, desde, 6-desde-hasta ) # 6-1-3 = 2 la prim vez. El 6 es la suma de los valores de cada torre 1+2+3
    print("Lleva la ficha ", colores[N], "desde",desde," hasta ", hasta )
    #Torres_Hanoi(N-1,6-desde-hasta, hasta )
    Torres_Hanoi(N-1, 6-desde-hasta  , hasta )


Torres_Hanoi(3, 1 , 3)

#torre auxiliar = 6 - desde - hasta

# mejora, que en el print salga que ficha se mueve:
# Ficha Roja la más peque
# Ficha Amarilla mediana
# Ficha verde grande

#llamadas recursivas:
# -> Torres_Hanoi(3, 1, 3)

# -> -> Torres_Hanoi(2, 1, 2)
# -> -> -> Torres_Hanoi(1, 1, 3)
# -> -> -> -> print(Lleva la ficha 1 hasta 3) (Amarilla//Verde 0 Roja)
# -> -> -> print(Lleva la ficha 1 hasta 2)
# -> -> -> Torres_Hanoi(1, 3, 2)
# -> -> -> -> print(Lleva la ficha 3 hasta 2)

# -> -> print(Lleva la ficha 1 hasta 3)

# -> -> Torres_Hanoi(2, 2, 3)
# -> -> -> Torres_Hanoi(1, 2, 1)
# -> -> -> -> print(Lleva la ficha 2 hasta 1)
# -> -> -> print(Lleva la ficha 2 hasta 3)
# -> -> -> Torres_Hanoi(1, 1, 3)
# -> -> -> -> print(Lleva la ficha 1 hasta 3)
```

```
#Sucesión_de_Fibonacci
#https://es.wikipedia.org/wiki/Sucesi%C3%B3n_de_Fibonacci
#Calculo del termino n-simo de la suscesión de Fibonacci
def Fibonacci(N:int):
  if N < 2:
    return 1
  else:
    return Fibonacci(N-1)+Fibonacci(N-2)

#Mejora complejidad, reduciendo a O(n)
def Fibonacci_lineal(N):
    if N < 2:
        return 1
    a, b = 1, 1
    for _ in range(2, N+1):
        a, b = b, a + b
    return b

print(Fibonacci(8))
print(Fibonacci_lineal(8))
```

##Devolución de cambio por técnica voraz

```
def cambio_monedas(N, SM):
  SOLUCION = [0]*len(SM)   #SOLUCION = [0,0,0,0,..]
  ValorAcumulado = 0

  for i,valor in enumerate(SM):
    monedas =  (N-ValorAcumulado)//valor
    SOLUCION[i] = monedas
    ValorAcumulado = ValorAcumulado + monedas*valor

    if ValorAcumulado == N:
      return SOLUCION


cambio_monedas(15,[25,10,5,1]) #primero monedas de mayor valor25<10<5<1
```

## N-Reinas por técnica de vueta atrás


```
#busca todas las soluciones posibles
def escribe(S): #para dibujar el tablero
  n = len(S)
  for x in range(n):
    print("")
    for i in range(n):
      if S[i] == x+1:
        print(" X " , end="")
      else:
        print(" - ", end="")


def es_prometedora(SOLUCION,etapa):

  print("*********")
  print(SOLUCION)
  escribe(SOLUCION)
  print("\n *********")

  #Si la solución tiene dos valores iguales no es valida => Dos reinas en la misma fila
  for i in range(etapa+1):
    #print("El valor " + str(SOLUCION[i]) + " está " +  str(SOLUCION.count(SOLUCION[i])) + " veces")
    if SOLUCION.count(SOLUCION[i]) > 1: # si por ejemplo la solucion es (0,2,2,1) no valido dos reinas en 2esto da >1
      return False

    #Verifica las diagonales
    for j in range(i+1, etapa +1 ):
      #print("Comprobando diagonal de " + str(i) + " y " + str(j))
      if abs(i-j) == abs(SOLUCION[i]-SOLUCION[j]) : return False
  return True



def reinas(N, solucion=[], etapa=0): #colocamos las reinas siguiendo el arbol de expamción(estructura de datos)
  if len(solucion) == 0:
      solucion=[0 for i in range(N)]

  for i in range(1, N+1): #la i es la fila donde las coloca
    solucion[etapa] = i

    if es_prometedora(solucion, etapa): # si no se sumple que es prometedora paramos la expansoin del arbol
      if etapa == N-1: #si es prometedora y es la ultima reina
        print(solucion) #finalicacion hemos encontrado solucion
        #escribe(solucion)
        print()
      else:
        reinas(N, solucion, etapa+1)
    else:
      None

    solucion[etapa] = 0

reinas(4) # para N = 4reinas
#(2,4,1,3)una solucion, se almacena
```

## Viaje por el rio. Programación dinámica

```
TARIFAS = [             #[nodo0,nodo1,nodo2,nodo3,nodo4,nodo5,nodo6,nodo7] sigue los pesos de las transaprencias de clase
[0,5,4,3,999,999,999],  #pesos de lo que cuesta ir a nodos conectados con nodo 0 representados en fila 0
[999,0,999,2,3,999,11], #destinos nodo 0 en la columna 0
[999,999, 0,1,999,4,10], #en la diagonal de la matriz de adyacencia hay 0 porque es el coste de ir de un nodo a sí mismo
[999,999,999, 0,5,6,9], #se ponen pesos grandes porque al ser tan gandes se ddescartarán esos caminos al querer miniizar el costo
[999,999, 999,999,0,999,4],
[999,999, 999,999,999,0,3],
[999,999,999,999,999,999,0]
]
################################################################
def Precios(TARIFAS):
################################################################
  #Total de Nodos
  N = len(TARIFAS[0]) #inicializamos totl nodos 7 ndoos

  #Inicialización de la tabla de precios a valores grandes
  PRECIOS = [ [9999]*N for i in [9999]*N]
  RUTA = [ [""]*N for i in [""]*N]

  for i in range(0,N-1):
    RUTA[i][i] = i             #Para ir de i a i se "pasa por i"
    PRECIOS[i][i] = 0          #Para ir de i a i se se paga 0
    for j in range(i+1, N):
      MIN = TARIFAS[i][j]
      RUTA[i][j] = i

      for k in range(i, j):
        if PRECIOS[i][k] + TARIFAS[k][j] < MIN:
            MIN = min(MIN, PRECIOS[i][k] + TARIFAS[k][j] )
            RUTA[i][j] = k          #Anota que para ir de i a j hay que pasar por k
        PRECIOS[i][j] = MIN

  return PRECIOS,RUTA
################################################################
PRECIOS,RUTA = Precios(TARIFAS)
#print(PRECIOS[0][6])

print("PRECIOS")
for i in range(len(TARIFAS)):
  print(PRECIOS[i])

print("\nRUTA")
for i in range(len(TARIFAS)):
  print(RUTA[i])

#Determinar la ruta con Recursividad
def calcular_ruta(RUTA, desde, hasta):
  if desde == hasta:
    #print("Ir a :" + str(desde))
    return ""
  else:
    return str(calcular_ruta( RUTA, desde, RUTA[desde][hasta])) +  \
                ',' + \
                str(RUTA[desde][hasta] \
              )

print("\nLa ruta es:")
calcular_ruta(RUTA, 0,6)
```
